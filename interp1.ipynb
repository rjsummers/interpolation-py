```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math

sns.set()
```

```python
f0 = 0.3817

xp = np.arange(32)
yp = np.sin((2 * np.pi * f0) * xp)

plt.scatter(xp, yp)
```

```python
x = np.arange(31, step=0.01)
y = np.zeros(np.shape(x))
```

```python
for ii in np.arange(np.size(x)):
    idx = int(round(x[ii]))
    y[ii] = yp[idx]

plt.plot(x, y)
plt.scatter(xp, yp)
```

```python
for ii in np.arange(np.size(x)):
    dec, idx = math.modf(x[ii])
    idx = int(idx)
    y[ii] = yp[idx] + (yp[idx + 1] - yp[idx]) * dec

plt.plot(x, y)
plt.scatter(xp, yp)
```
